```python
import pandas as pd

df1=pd.read_csv('dataset.csv')
des=pd.read_csv('symptom_Description.csv')
pre=pd.read_csv('symptom_precaution.csv')
```

```python
df1.head()
```

```python
df1['Symptoms'] = df1.iloc[:, 1:17].apply(lambda x: list(x.dropna()), axis=1)
df1
```

```python
column='Disease'
df_new2 = df1.drop_duplicates(subset=column)
final_df = df_new2[['Disease','Symptoms']]
final_df
```

```python
pre
```

```python
pre['Precautions'] = pre.iloc[:, 1:4].apply(lambda x: list(x.dropna()), axis=1)
pre
```

```python
final_df1 = pre[['Disease','Precautions']]
final_df1
```

```python
merged_df1 = pd.merge(final_df, final_df1, on='Disease')
merged_df1 = merged_df1[['Disease', 'Symptoms', 'Precautions']]
merged_df1
```

```python
merged_df2 = pd.merge(merged_df1, des, on='Disease')
merged_df2 = merged_df2[['Disease', 'Symptoms', 'Precautions' ,'Description']]
merged_df2
```
